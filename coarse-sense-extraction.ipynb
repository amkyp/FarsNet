```python
import json

# Read the JSON file
with open('senses_for_words_20240621151624.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

for key, value in data.items():
    result = []
    for sense in value:
        sense_info = {
            'Meaning': sense['meaning'],
            'Example': sense['example'],
            'POS': sense['pos'],
            'Semantic Category': sense['semantic_category'],
            'Sense ID': sense['sense_id'],
            'Synset ID': sense['synset_id']
        }
        result.append(sense_info)
    json_result = json.dumps(result, indent=4, ensure_ascii=False)
    print(f"Word: {key}")
    print(json_result)
    break
```
